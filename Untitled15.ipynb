```python
# Load the data into a Pandas DataFrame
data = pd.read_excel("/content/data texture final.xlsx")
```

```python
# Prepare data by dropping non-relevant columns
data.drop(['patient number', 'ROI number', 'seg', 'pacs', 'age', 'sex 1:male 2:female', 'E-L pvp'], axis=1, inplace=True)

# Split data into input (X) and output (y) variables
X = data.iloc[:, 1:].values
y = data.iloc[:, 0].values
```

```python
# Shift the values to make them non-negative
X -= X.min()
```

```python
from scipy.stats import mstats

# Replace extreme values in the 'data' array with the nearest values within the 5th and 95th percentile range
data_winsorized = mstats.winsorize(X, limits=[0.05, 0.05])
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X -= X.min()
```

```python
import pandas as pd

# Convert X to a pandas DataFrame
df = pd.DataFrame(X)

# Calculate the correlation matrix
corr_matrix = df.corr().abs()

# Find the highly correlated features
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]

# Drop the highly correlated features
df_filtered = df.drop(to_drop, axis=1)

# Convert back to a numpy array
X_filtered = df_filtered.to_numpy()
```

```python
print(X_filtered.shape)
```

```python
X = X_filtered
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Instantiate a PCA object
pca = PCA()

# Fit the PCA object to the data and transform the data
X_pca = pca.fit_transform(X)

# Compute the cumulative explained variance ratio
cumulative_var = np.cumsum(pca.explained_variance_ratio_)

# Plot the cumulative explained variance ratio versus the number of components
plt.plot(range(1, len(cumulative_var) + 1), cumulative_var)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('PCA Cumulative Explained Variance Ratio')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Fit PCA with all features
pca = PCA().fit(X)

# Find the number of components that explain 95% of the variance
variance_ratio_cumsum = np.cumsum(pca.explained_variance_ratio_)
num_components = np.argmax(variance_ratio_cumsum >= 0.95) + 1

# Fit PCA with the selected number of components
pca = PCA(n_components=num_components)
X_pca = pca.fit_transform(X)
```

```python
print(X_pca.shape)
```

```python
X = X_pca
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
classifiers = {
    "SVC": SVC(kernel='linear', probability=True, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=10, random_state=0),
    "Naive Bayes 1": GaussianNB(),
    "Naive Bayes 2": GaussianNB(var_smoothing=1e-9),
    "KNN 1": KNeighborsClassifier(n_neighbors=3, metric='euclidean'),
    "KNN 2": KNeighborsClassifier(n_neighbors=5, metric='euclidean'),
    "KNN 3": KNeighborsClassifier(n_neighbors=7, metric='euclidean'),
    "Neural Network": MLPClassifier(hidden_layer_sizes=(50,50), max_iter=1000, activation='relu', solver='adam', random_state=1)
}
```

```python
#@title Default title text
for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_train)
    accuracy = accuracy_score(y_train, y_pred)
    precision = precision_score(y_train, y_pred)
    recall = recall_score(y_train, y_pred)
    f1 = f1_score(y_train, y_pred)
    print("{} evaluation metrics on train set:".format(name))
    print("Accuracy: {:.2f}%".format(accuracy*100))
    print("Precision: {:.2f}%".format(precision*100))
    print("Recall: {:.2f}%".format(recall*100))
    print("F1-score: {:.2f}%".format(f1*100))
    print(classification_report(y_train, y_pred))
```

```python
# Initialize lists to store the ROC curve data
train_fpr_list = []
train_tpr_list = []
test_fpr_list = []
test_tpr_list = []

# Loop through the classifiers
for clf_name, clf in classifiers.items():
    # Train the classifier
    clf.fit(X_train, y_train)
    
    # Get predicted probabilities for the positive class (1)
    train_y_proba = clf.predict_proba(X_train)[:, 1]
    test_y_proba = clf.predict_proba(X_test)[:, 1]
    
    # Calculate false positive rate, true positive rate, and thresholds for the train set
    train_fpr, train_tpr, train_thresholds = roc_curve(y_train, train_y_proba)
    
    # Calculate false positive rate, true positive rate, and thresholds for the test set
    test_fpr, test_tpr, test_thresholds = roc_curve(y_test, test_y_proba)
    
    # Append the FPR and TPR data to the lists
    train_fpr_list.append(train_fpr)
    train_tpr_list.append(train_tpr)
    test_fpr_list.append(test_fpr)
    test_tpr_list.append(test_tpr)
    
    # Calculate AUC score
    auc_train = roc_auc_score(y_train, train_y_proba)
    auc_test = roc_auc_score(y_test, test_y_proba)
    
    # Print the evaluation metrics
    print(f"{clf_name}: Train AUC = {auc_train:.3f}, Test AUC = {auc_test:.3f}")
    
# Plot the ROC curves for the train set
plt.figure(figsize=(10,8))
for i in range(len(classifiers)):
    plt.plot(train_fpr_list[i], train_tpr_list[i], label=list(classifiers.keys())[i])
    
# Plot the ROC curves for the test set
for i in range(len(classifiers)):
    plt.plot(test_fpr_list[i], test_tpr_list[i], linestyle='--', label=list(classifiers.keys())[i] + ' (test)')
    
# Plot the random guess line
plt.plot([0, 1], [0, 1], color='black', linestyle='--')

# Set the x and y axis labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')

# Add the legend to the plot
plt.legend(loc='lower right')

# Show the plot
plt.show()
# save as PDF
plt.savefig("AUCS.pdf", bbox_inches="tight")

# save as image in high resolution
plt.savefig("AUCS.png", dpi=300, bbox_inches="tight")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from sklearn.metrics import roc_curve, roc_auc_score, recall_score, confusion_matrix
from sklearn.model_selection import train_test_split


# Initialize lists to store the ROC curve data
train_fpr_list = []
train_tpr_list = []
test_fpr_list = []
test_tpr_list = []
train_sens_list = []
train_spec_list = []
test_sens_list = []
test_spec_list = []

# Loop through the classifiers
for clf_name, clf in classifiers.items():
    # Train the classifier
    clf.fit(X_train, y_train)
    
    # Get predicted probabilities for the positive class (1)
    train_y_proba = clf.predict_proba(X_train)[:, 1]
    test_y_proba = clf.predict_proba(X_test)[:, 1]
    
    # Calculate false positive rate, true positive rate, and thresholds for the train set
    train_fpr, train_tpr, train_thresholds = roc_curve(y_train, train_y_proba)
    
    # Calculate false positive rate, true positive rate, and thresholds for the test set
    test_fpr, test_tpr, test_thresholds = roc_curve(y_test, test_y_proba)
    
    # Append the FPR and TPR data to the lists
    train_fpr_list.append(train_fpr)
    train_tpr_list.append(train_tpr)
    test_fpr_list.append(test_fpr)
    test_tpr_list.append(test_tpr)
    
    # Calculate AUC score
    auc_train = roc_auc_score(y_train, train_y_proba)
    auc_test = roc_auc_score(y_test, test_y_proba)
    
    # Calculate sensitivity/recall score
    train_recall = recall_score(y_train, clf.predict(X_train))
    test_recall = recall_score(y_test, clf.predict(X_test))
    
    # Calculate specificity score
    train_tn, train_fp, train_fn, train_tp = confusion_matrix(y_train, clf.predict(X_train)).ravel()
    train_specificity = train_tn / (train_tn + train_fp)
    test_tn, test_fp, test_fn, test_tp = confusion_matrix(y_test, clf.predict(X_test)).ravel()
    test_specificity = test_tn / (test_tn + test_fp)
    
    # Print the evaluation metrics
    print(f"{clf_name}: Train AUC = {auc_train:.3f}, Test AUC = {auc_test:.3f}, Train Sensitivity = {train_recall:.3f}, Test Sensitivity = {test_recall:.3f}, Train Specificity = {train_specificity:.3f}, Test Specificity = {test_specificity:.3f}")
    
    # Append the sensitivity and specificity data to the lists
    train_sens_list.append(train_recall)
    train_spec_list.append(train_specificity)
    test_sens_list.append(test_recall)
    test_spec_list.append(test_specificity)

# Plot the ROC curves for the train and test sets
plt.plot([0, 1], [0, 1], linestyle='--', color='grey')
for train_fpr, train_tpr, test_fpr, test_tpr in zip(train_fpr_list, train_tpr_list, test_fpr_list, test_tpr_list):
    plt.plot(train_fpr, train_tpr, color='blue', label='Train')
```

```python
from sklearn.metrics import auc
import numpy as np
from sklearn.utils import resample

# Initialize lists to store the ROC curve data
train_fpr_list = []
train_tpr_list = []
test_fpr_list = []
test_tpr_list = []
train_auc_bootstraps = []
test_auc_bootstraps = []

# Loop through the classifiers
for clf_name, clf in classifiers.items():
    # Train the classifier
    clf.fit(X_train, y_train)
    
    # Get predicted probabilities for the positive class (1)
    train_y_proba = clf.predict_proba(X_train)[:, 1]
    test_y_proba = clf.predict_proba(X_test)[:, 1]
    
    # Calculate false positive rate, true positive rate, and thresholds for the train set
    train_fpr, train_tpr, train_thresholds = roc_curve(y_train, train_y_proba)
    
    # Calculate false positive rate, true positive rate, and thresholds for the test set
    test_fpr, test_tpr, test_thresholds = roc_curve(y_test, test_y_proba)
    
    # Append the FPR and TPR data to the lists
    train_fpr_list.append(train_fpr)
    train_tpr_list.append(train_tpr)
    test_fpr_list.append(test_fpr)
    test_tpr_list.append(test_tpr)
    
    # Calculate AUC score
    auc_train = roc_auc_score(y_train, train_y_proba)
    auc_test = roc_auc_score(y_test, test_y_proba)
    
    # Calculate bootstrap confidence intervals for the AUC score
    for i in range(1000):
        train_y_sample = resample(train_y_proba, replace=True, n_samples=len(train_y_proba))
        test_y_sample = resample(test_y_proba, replace=True, n_samples=len(test_y_proba))
        train_auc_bootstraps.append(auc(train_fpr, train_tpr))
        test_auc_bootstraps.append(auc(test_fpr, test_tpr))
    train_auc_lower, train_auc_upper = np.percentile(train_auc_bootstraps, (2.5, 97.5))
    test_auc_lower, test_auc_upper = np.percentile(test_auc_bootstraps, (2.5, 97.5))
    
    # Print the evaluation metrics
    print(f"{clf_name}: Train AUC = {auc_train:.3f} ({train_auc_lower:.3f}-{train_auc_upper:.3f}), Test AUC = {auc_test:.3f} ({test_auc_lower:.3f}-{test_auc_upper:.3f})")
```
